### Importing Libaries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from keras.models import load_model
model = load_model('trained_traffic_model.h5')
```

```python
image='test_traffic_1.jfif'
img = cv2.imread(image,cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
img = cv2.equalizeHist(img)
img = img/255
img = cv2.resize(img,(60,60))

img = img.reshape(1,60,60,1)
predict_x=model.predict(img) 
classIndex=np.argmax(predict_x,axis=1)

if classIndex == 0:
    text='20 km/h'
if classIndex == 1:
    text='50 km/h'
if classIndex == 2:
    text='30 km/h'
if classIndex == 3:
    text='80 km/h'
if classIndex == 4:
    text='TURN LEFT'
if classIndex == 5:
    text='TURN RIGHT'
if classIndex == 6:
    text='100 km/h'
if classIndex == 7:
    text='GO STRAIGHT'
if classIndex == 8:
    text='NO HORN'
    
```

### Displaying Class of given Images

```python
 print(text)
```

###  Displaying Image 

```python
img=cv2.imread(image)
while True:
    cv2.imshow("image",img)
    cv2.rectangle(img,(15,15),(250,60),(255,50,50),-1)
    cv2.putText(img,text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
